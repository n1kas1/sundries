```python
games["user_score"] = pd.to_numeric(games.user_score, errors='coerce')
games.head()
```

```python
games["sum_sales"] = games.na_sales + games.eu_sales + games.jp_sales + games.other_sales
```

```python
games.head()
```

```python
games_by_year = games.groupby('year_of_release').size()

plt.plot(games_by_year.index, games_by_year.values, marker='.')
plt.title('Количество игр по годам')
plt.xlabel('Год')
plt.ylabel('Количество игр')
plt.grid(True)
plt.show()
```

Видим, что данные до 1994 года содержат мало информации поэтому вообще можно было бы их убрать из рассмотрения - сделаем это

```python
new_games = games[games.year_of_release > 1994] # будем его использовать где-то в дальнейшем
```

```python
platform_sales = games.groupby('platform')['sum_sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
platform_sales.head(10).plot(kind='bar')
plt.title('Суммарные продажи по платформам')
plt.xlabel('Платформа')
plt.ylabel('Суммарные продажи (млн)')
plt.xticks(rotation=45)
plt.show()
```

```python
top_platforms = platform_sales.head(10).index
platform_year_sales = games[games['platform'].isin(top_platforms)].groupby(['year_of_release', 'platform'])['sum_sales'].sum().unstack(fill_value=0)

plt.figure(figsize=(15, 8))
platform_year_sales.plot(kind='line', marker='o')
plt.title('Динамика продаж по годам для топ-10 платформ')
plt.xlabel('Год')
plt.ylabel('Суммарные продажи (млн)')
plt.legend(title='Платформа', loc='upper left')
plt.grid(True)
plt.show()
```

* Имеем, что популярные платформы имеют жизненный цикл около 6-10 лет.
* Новые платформы появляются примерно каждые 5-7 лет, которые как раз заменяют старые
* PS2, XBOX360, блыи доминирующими в 2000-х, но к 2016 году их популярность сильно снизилась

```python
actual_period = games[games['year_of_release'] >= 2010]
```

```python
top_platforms_actual = actual_period.groupby('platform')['sum_sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
top_platforms_actual.plot(kind='bar')
plt.title('Суммарные продажи по платформам (2010-2016)')
plt.xlabel('Платформа')
plt.ylabel('Суммарные продажи (млн)')
plt.xticks(rotation=45)
plt.show()
```

```python
platform_year_sales_actual = actual_period[actual_period['platform'].isin(top_platforms_actual.index)].groupby(['year_of_release', 'platform'])['sum_sales'].sum().unstack(fill_value=0)

plt.figure(figsize=(15, 8))
platform_year_sales_actual.plot(kind='line', marker='.')
plt.title('Динамика продаж по годам для топ-10 платформ (2010-2016)')
plt.xlabel('Год')
plt.ylabel('Суммарные продажи (млн)')
plt.legend(title='Платформа', bbox_to_anchor=(1, 1), loc='upper left')
plt.grid(True)
plt.show()
```

* Имеем, что PS4 и XOne показывают рост (с момента своего запуска)
* 3DS - просто стабильные продажи
* WiiU - не достигло нормальных продаж

Имеем, потенциально прибыльные платформы: 
- PS4 - самая популярная с быстрым ростом
- XOne - вторая по популярности
- 3DS - стабильный "друг"

```python
top_5_platforms = top_platforms_actual.head(5).index
data_for_boxplot = actual_period[actual_period['platform'].isin(top_5_platforms)]

plt.figure(figsize=(12, 6))
sns.boxplot(data=data_for_boxplot, x='platform', y='sum_sales')
plt.title('Распределение глобальных продаж по топ-5 платформам')
plt.xlabel('Платформа')
plt.ylabel('Глобальные продажи (млн)')
plt.yscale('log')  # лучше выглядит
plt.show()
```

* PS4, PS3 и X360 имеют высокие медианные значения продаж и большое количество выбросов
* 3DS имеет более низкие медианные продажи, но стабильное распределение без сильных выбросов

```python
ps4_data = actual_period[(actual_period['platform'] == 'PS4') & 
                         (actual_period['critic_score'].notna()) & 
                         (actual_period['user_score'].notna())]

# Построим диаграмму рассеяния для оценок критиков
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(ps4_data['critic_score'], ps4_data['sum_sales'], alpha=0.5)
plt.title('Влияние оценки критиков на продажи (PS4)')
plt.xlabel('Оценка критиков')
plt.ylabel('Глобальные продажи (млн)')
plt.grid(True)

# Построим диаграмму рассеяния для оценок пользователей
plt.subplot(1, 2, 2)
plt.scatter(ps4_data['user_score'], ps4_data['sum_sales'], alpha=0.5)
plt.title('Влияние оценки пользователей на продажи (PS4)')
plt.xlabel('Оценка пользователей')
plt.ylabel('Глобальные продажи (млн)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
genre_sales = actual_period.groupby('genre')['sum_sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
genre_sales.plot(kind='bar')
plt.title('Суммарные продажи по жанрам (2010-2016)')
plt.xlabel('Жанр')
plt.ylabel('Суммарные продажи (млн)')
plt.xticks(rotation=45)
plt.show()
```

```python
na_top_platforms = actual_period.groupby('platform')['na_sales'].sum().sort_values(ascending=False).head(5)
print('Топ-5 платформ в Северной Америке:')
print(na_top_platforms)
```

```python
eu_top_platforms = actual_period.groupby('platform')['eu_sales'].sum().sort_values(ascending=False).head(5)
print('\nТоп-5 платформ в Европе:')
print(eu_top_platforms)
```

```python
jp_top_platforms = actual_period.groupby('platform')['jp_sales'].sum().sort_values(ascending=False).head(5)
print('\nТоп-5 платформ в Японии:')
print(jp_top_platforms)
```

```python
rating_sales = actual_period.groupby('rating')[['na_sales', 'eu_sales', 'jp_sales']].sum()

plt.figure(figsize=(12, 6))
rating_sales.plot(kind='bar', stacked=True)
plt.title('Влияние рейтинга ESRB на продажи в разных регионах')
plt.xlabel('Рейтинг ESRB')
plt.ylabel('Суммарные продажи (млн)')
plt.xticks(rotation=45)
plt.legend(title='Регион')
plt.show()
```
